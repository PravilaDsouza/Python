```python

class ExcelUploader:
    def __init__(self, root):
        self.root = root
        self.root.title("Excel File Uploader")
        self.root.geometry("700x300+400+150")

        self.file_path_label = tk.Label(root, text="Choose File",font=('Verdana', 14, 'bold'))
        self.file_path_label.place(x=50,y=50)
        
        self.entry_path = Entry(self.root, font=('Verdana', 14))
        self.entry_path.place(x=180, y=50)

        self.browse_button = tk.Button(root, text="Browse", command=self.browse_file, height=1,width=10)
        self.browse_button.place(x=440, y=50)

        self.submit_button = tk.Button(root, text="Submit", command=self.submit_file, height=2,width=10)
        self.submit_button.place(x=280, y=150)

    def browse_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("Excel files", "*.xlsx;*.xls")])
        if file_path:
            self.entry_path.delete(0,tk.END)
            self.entry_path.insert(0,file_path)
            
            
    def submit_file(self):
        file_path=self.entry_path.get()
        if file_path:
            try:
                df = pd.read_excel(file_path)
                print(df)
                messagebox.showinfo("Success", "File submitted successfully.")
                
                subprocess.run(["Start","excel",file_path],shell=True)
            except pd.errors.EmptyDataError:
                messagebox.showwarning("Warning", "The selected Excel file is empty.")
                
            except PermissionError:
                messagebox.showerror("Error","Permission Denied: Check file Permissions.")
            except Exception as e:
                messagebox.showerror("Error", f"Error reading Excel file: {e}")
        else:
            messagebox.showwarning("Warning", "Please select an Excel file first.")


if __name__ == "__main__":
    root = tk.Tk()
    app = ExcelUploader(root)
    root.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog
from tkinter.messagebox import *
import openpyxl
import mysql.connector

class ExcelUploader:
    def __init__(self, root):
        self.root = root
        self.root.title("Excel To MySql")
        self.root.geometry("700x300+400+150")
        
       # Create a frame to hold the button
        self.frame = tk.Frame(root)
        self.frame.pack(expand=True)

        # Create the button and center it in the frame
        self.select_button = tk.Button(self.frame, text="Select Excel File", height=2, width=15, font=('Verdana', 14, 'bold'), command=self.extract_and_insert_data)
        self.select_button.pack(side="left", anchor="center")
        

    def extract_and_insert_data(self):
    # Ask user to select an Excel file
        file_path = filedialog.askopenfilename(filetypes=[("Excel Files", "*.xlsx;*.xls")])
    
        if not file_path:
            return  # User canceled the file selection

        try:
        # Connect to the MySQL database
            db_connection = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Mysql@123',   
                database='excel'
            )

            # Open the Excel file
            workbook = openpyxl.load_workbook(file_path)

            # Select a specific sheet (change 'Sheet1' to the name of your sheet)
            sheet = workbook.active

            # Initialize an empty list to store your data
            data_array = []

            # Iterate through rows in the sheet and store data in the list
            for row in sheet.iter_rows(values_only=True):
                data_array.append(row)

            # Close the Excel file
            workbook.close()

            # Create a cursor to interact with the database
            cursor = db_connection.cursor()

            # Define an SQL INSERT statement
            insert_query = "INSERT INTO work (col1, col2, col3, col4) VALUES (%s, %s, %s, %s)"

            # Iterate through the data and insert it into the database
            for row in data_array:
                cursor.execute(insert_query, row)

            # Commit the changes to the database
            db_connection.commit()

            # Close the cursor and database connection
            cursor.close()
            db_connection.close()
            
            #create a message box to display the message
            messagebox.showinfo("Success","Data extracted and inserted successfully!")

        #create a message box to display error message
        except Exception as e:
            messagebox.showerror("Error",f"Error: {str(e)}")

#create the main tkinter window and instantiate the ExcelUploader class with the main window as its parent
if __name__ == "__main__":
    root = tk.Tk()
    app = ExcelUploader(root)
    #start the tkinter event loop
    root.mainloop()
```
